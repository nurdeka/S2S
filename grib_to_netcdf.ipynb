```python
import pandas as pd
import xarray as xarray
```

```python
input_file = 'output_'
output_file = 'netcdf_'
```

```python
now = pd.Timestamp(year=2015, month=2, day=1)
L = pd.period_range(now, freq='M', periods=(12*8)).strftime('%Y%m').tolist()
#print (L)
for i in L:
    data = xarray.open_dataset(input_file+i, engine='cfgrib')
    data.to_netcdf(output_file+i+'.nc')
    print('done'+output_file+i+'.nc')
```
