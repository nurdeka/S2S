```python
import pandas as pd
import xarray as xr
nc_file = "blend_1991-present_0.5.nc"
NC = xr.open_dataset(nc_file, decode_times=False)
```

```python
import numpy as np
import pandas as pd
grid = []
for lat in np.arange(-11,5.5,0.1):
    for lon in np.arange(95.05,140.6,0.1):
        grid.append([lat, lon])

df_grid = pd.DataFrame(grid, columns=['latitude', 'longitude'])
df_grid.head()
```

```python
lat = df_grid.latitude.tolist()
lon = df_grid.longitude.tolist()
```

```python
dsloc = NC.sel(lat=0,lon=101,method='nearest')
```

```python
dsloc['precip'][:]
```

```python
Newdf = pd.DataFrame([])

for i,j in zip(lat, lon):
    dsloc = NC.sel(lat=i,lon=j,method='nearest')
    DT=dsloc.to_dataframe()

    # insert the name with your preferred column title:
    #DT.insert(loc=0,column="station",value=id)
    Newdf=Newdf.append(DT,sort=True)

print(Newdf.head())
```
